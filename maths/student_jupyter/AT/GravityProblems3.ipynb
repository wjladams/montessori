```python
from sympy import Symbol, N
from sympy.solvers import solve
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import Scatter, Figure, Layout
init_notebook_mode(connected=True)
```

```python
t = Symbol('t')
```

Working on **problem 4**

```python
solns=solve(-16*t**2+64*t+20)
solns
```

```python
[N(s) for s in solns]
```

```python
time_hit = N(solns[0])
time_hit
```

```python
x = 100 * time_hit
x
```

```python
# Graphing
x = lambda t: 100*t
y = lambda t: -16*t**2 +64*t+20
times = [t/100 for t in range(429)]
iplot([{"x":[x(t) for t in times], "y":[y(t) for t in times]}])
```
